```python

class Streamer:
    def __init__(self, nombre, plataforma, edad):
        self.nombre = nombre
        self.plataforma = plataforma
        self.edad = edad

    def iniciar_directo(self):
        print(f"¡{self.nombre} está EN VIVO en {self.plataforma}! '¡Hola chat, bienvenidos!'")

    def leer_donaciones(self):
        print(f"{self.nombre} dice: '¡Gracias por esos bits! Se agradece mucho el apoyo.'")

    def jugar(self):
        print(f"{self.nombre} está concentrado jugando. '¡No puede ser, me eliminaron!'")

streamer1 = Streamer("Samy Rivers", "Twitch", 25)
streamer2 = Streamer("Ibai Llanos", "YouTube", 28)

print("--- ESTADO DEL STREAMER 1 ---")
print(f"Nombre: {streamer1.nombre} | Plataforma: {streamer1.plataforma} | Edad: {streamer1.edad}")
streamer1.iniciar_directo()
streamer1.jugar()

print("\n--- ESTADO DEL STREAMER 2 ---")
print(f"Nombre: {streamer2.nombre} | Plataforma: {streamer2.plataforma} | Edad: {streamer2.edad}")
streamer2.iniciar_directo()
streamer2.leer_donaciones()

print("Programa finalizado")
print("Programa realizado por Zavala Medina NL= 54")
```
